```python
from pathlib import Path
import json
from collections import Counter

```

```python
ann_dir = Path("/home/the/Codes/Melanoma-detection/ds/ann")
results = []

def get_class_titles(json_path: Path):
    try:
        with json_path.open("r", encoding="utf-8") as f:
            data = json.load(f)
        return [o.get("classTitle") for o in data.get("objects", []) if o.get("classTitle")]
    except (json.JSONDecodeError, OSError) as e:
        print(f"Skip {json_path.name}: {e}")
        return []

# Example A: Print class titles per file
for p in ann_dir.glob("*.json"):
    titles = get_class_titles(p)
    if titles:
        base_id = Path(p.stem).stem  # handles ".jpg.json" -> "ISIC_0024306"
        results.append({"id": base_id,"titles": titles})
        
for row in results[:5]:
    print(row["id"],row["titles"])

# Example B: Count class titles across all files
counts = Counter()
for p in ann_dir.glob("*.json"):
    counts.update(get_class_titles(p))
print("\nTop classes:", counts.most_common(10))
```
